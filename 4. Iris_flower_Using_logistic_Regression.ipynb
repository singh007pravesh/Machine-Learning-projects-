```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df= pd.read_csv('iris.data' , header= None)
```

```python
col_name=[ 'sepal length','sepal width','petal length','petal width','class' ]
```

```python
df.columns=col_name
```

```python
df.head()
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x=df['sepal length'],y=df['sepal width'],kind='reg')
```

```python
sns.jointplot(x=df['petal length'],y=df['petal width'],kind='reg')
```

```python
sns.stripplot(x=df['sepal length'],y=df['petal width'],data=df)
```

```python
sns.stripplot(x=df['petal length'],y=df['petal width'],data=df)
```

```python
x = df.iloc[:,:4]
x.head()
```

```python
y = df.iloc[:,4:]
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test , y_train,y_test= train_test_split(x,y,test_size=0.4,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
pred
```

```python
y_test.head()
```

```python
model.score(x_test,y_test)
```

### Confusion Matrix.................

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
confusion_matrix(y_test,pred)
```

```python
accuracy_score(y_test,pred)
```
